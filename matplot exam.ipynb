```python
import matplotlib.pyplot as plt

import numpy as np

x=np.array([0.0,1.0,2.0,3.0])

y=np.array([3,8,1,10])

plt.plot(x,y,color="r",ls="dotted",marker ="o")

plt.show()
```

```python
import matplotlib.pyplot as  plt

import numpy as np

y=np.array([13, 20,24,45])

mylabels=["python", "chemistry", "english", "maths"]

mycolors=['blue','green','red','yellow']

myexplode=[0.2,0,0,0]

plt.pie(y,labels=mylabels,explode=myexplode,colors=mycolors)

plt.show()
```
